### Use an interpolation polynomial to appoximate the curve of bird's top portion.


```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
```

```python
def lagrange(x, data_x):
    n = len(data_x)
    L = np.empty(n)
    for i in range(n):
        mult = 1
        for j in range(n):
            if j != i:
                mult *= (x - data_x[j])/(data_x[i] - data_x[j])
        L[i] = mult
    return L

def interpolate(x, data_x, data_y):
    L = lagrange(x, data_x)
    f = data_y
    return np.sum(L*f)
```

```python
img_name = 'workshop_11_img.png'
img = plt.imread(img_name)
fig, ax = plt.subplots()
ax.imshow(img, extent=[0, img.shape[1], 0, img.shape[0]])
ax.grid()
fig.canvas.toolbar_visible = False
fig.canvas.header_visible = False
fig.canvas.resizable = False
```

```python
data_x = [87.3, 135.7, 200.3, 255.1, 343.8, 392.2, 430.9, 482.5, 534.1]
data_y = [366.6, 339.2, 318.2, 323.1, 371.5, 388.8, 398.9, 404.7, 411.2]

sample = 100
inter_x = np.linspace(min(data_x), max(data_x), sample)
inter_y = np.empty(sample)
for i in range(sample):
    inter_y[i] = interpolate(inter_x[i], data_x, data_y)
    
img = plt.imread(img_name)
fig, ax = plt.subplots()
ax.imshow(img, extent=[0, img.shape[1], 0, img.shape[0]])
ax.grid()

ax.scatter(data_x, data_y, color='blue')
ax.plot(inter_x, inter_y, color='red')

fig.canvas.toolbar_visible = False
fig.canvas.header_visible = False
fig.canvas.resizable = False
```

## Problem 2

For your desired figure, use interpolation method to appoximate the curve of its portions.

```python
img_name = 'plane.png'

img = plt.imread(img_name)
fig, ax = plt.subplots()
ax.imshow(img, extent=[0, 512, 0, 512]) # width and length of image
ax.grid()
fig.canvas.toolbar_visible = True
fig.canvas.header_visible = False
fig.canvas.resizable = False
```

```python
data_x = [15.5, 34, 55, 86, 110, 112]
data_y = [395.5, 396, 391.96, 333.93, 296, 294.8]
data_xnew = [112, 130, 155, 200, 250, 300, 346, 373, 405, 440, 455,472.5]
data_ynew = [294.8, 294.8, 294.8, 294.8, 294.8, 294.8, 294.8, 288, 262, 242, 238.9, 238]

sample = 100
inter_x = np.linspace(min(data_x), max(data_x), sample)
inter_y = np.empty(sample)
inter_xnew = np.linspace(min(data_xnew), max(data_xnew), sample)
inter_ynew = np.empty(sample)

for i in range(sample):
    inter_y[i] = interpolate(inter_x[i], data_x, data_y)
    inter_ynew[i] = interpolate(inter_xnew[i], data_xnew, data_ynew)
    
img = plt.imread(img_name)
fig, ax = plt.subplots()
ax.imshow(img, extent=[0, img.shape[1], 0, img.shape[0]])
ax.grid()

ax.scatter(data_x, data_y, color='blue')
ax.plot(inter_x, inter_y, color='red')
ax.scatter(data_xnew, data_ynew, color='blue')
ax.plot(inter_xnew, inter_ynew, color='red')

fig.canvas.toolbar_visible = True
fig.canvas.header_visible = False
fig.canvas.resizable = False
```

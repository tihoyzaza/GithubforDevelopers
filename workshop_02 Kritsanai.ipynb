## PROBLEM

กําหนดฟังก์ชัน

$$ f(x) = x^3 - 6x^2 + 11x - 5.9 $$

จงหาค่ารากของสมการโดยวิธีกราฟ (Graphical Method)

```python
# Solution here ...

from matplotlib import pyplot as plt

x = [-4, 0, 8, 15]
y = [2, 0, -2, 10]

plt.plot(x, y)
plt.grid()
```

```python
def f(x):
    return (x**3) - (6*x**2) + (11*x) - (5.9)


x = 10
y = f(x)

print(f'x = {x} --> y =f({x}) =  {f(x)}')
      
xlist = [-5, -2, 0, 2, 5, 10]
ylist = [0] * 6 

for  k in range(6):
      ylist[k] = f(xlist[k])
      
print(xlist)
print(ylist)

plt.plot(xlist, ylist)
plt.grid()
```

```python
N = 100
xmin = 0.953
xmax = 0.954

xlist = [0] * N
ylist = [0] * N

for k in range(N):
    x = xmin + (xmax - xmin) * k/(N-1)
    xlist[k] = x
    y = f(x)
    ylist[k]  = y


plt.figure(figsize = (10,3))
plt.plot(xlist, ylist)
plt.grid()
```

## PROBLEM

กําหนดฟังก์ชัน

$$ f(x) = x^3 - 6x^2 + 11x - 5.9 $$

จงหาค่ารากของสมการโดยวิธีแบ่งครึ่งช่วง (Bisection Method)

กำหนดค่าเริ่มต้น
    $$ x_l = 2.5 \quad\text{ และ }\quad x_u = 3.5  $$
โดยที่ค่าร้อยละของความคลาดเคลื่อนสัมพัทธ์ คือ $\varepsilon_{s} = 0.1 \%$

```python
xl = 2.5
xu = 3.5
xr = (xl + xu) / 2

print(xr)

ea = 100

while ea > 0.1:

    def f(x):
        return (x**3) - (6*x**2) + (11*x) - (5.9)

    if f(xl)*f(xr) > 0:
        xl = xr
        xrn = (xl + xu) / 2
        ea = (abs(xrn - xr) / abs(xrn)) * 100
        xr = xrn
        print(f'percent = {ea}')
        print(f' solution = {xrn}')

    elif f(xl)*f(xr) < 0:
        xu = xr
        xrn = (xl + xu) / 2
        ea = (abs(xrn - xr) / abs(xrn)) * 100
        xr = xrn
        print(f'percent = {ea}')
        print(f' solution = {xrn}')
```

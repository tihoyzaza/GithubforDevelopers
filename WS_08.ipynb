## Problem 01

#### A) Find the least square regression line for the following set of data

| $i$ | 1 | 2 | 3 | 4 | 5 |
|---|------|------|------|-------|-------|
| $x_i$ | 2.10 | 6.22 | 7.17 | 10.52 | 13.68 |
| $y_i$ | 2.90 | 3.83 | 5.98 | 5.71  | 7.74  |

````python
import numpy as np
import matplotlib.pyplot as plt


data_x = np.array([2.10, 6.22, 7.17, 10.52, 13.68])
data_y = np.array([2.90, 3.83, 5.98, 5.71, 7.74])
plt.scatter(data_x, data_y) #plot dot โดยไม่มีการลากเส้น
plt.xlabel('x-value')
plt.ylabel('y-value')
plt.grid()
plt.show()
````

#### Linear Regression Model
$$ \hat{Y} = XB = \begin{bmatrix} 1 & X_1 \\ 1 & X_2 \\ \vdots & \vdots \\ 1 & X_n \end{bmatrix}\begin{bmatrix} B_0 \\ B_1 \end{bmatrix}, $$
where 
$$ B = (X^{T}X)^{-1}X^{T}Y $$
$$ \hat{Y} = B_0 + B_1x$$

````python
n = data_x.shape[0] # [0, 0, 0, 0, 0]
ones = np.ones((n, 1)) # [1, 1, 1, 1, 1] แนวตั้ง
X = np.concatenate((ones, data_x.reshape(n,1)), axis=1) # นำ array มารวมกัน example 
Y = data_y.reshape(n, 1)

print(f'X = \n {X}')
print(f'Y = \n {Y}')
````

````python
#from numpy.linalg import inv
B = np.linalg.inv(X.T @ X) @ X.T @ Y #faction find invert matrix
print(B)
````

````python
print(f'yhat = {B[0]} + {B[1]} * x')
````

````python
def model_p1(x):
    yhat = B[0] + B[1]*x
    return yhat
````

````python
yhat = model_p1(data_x)

plt.scatter(data_x, data_y)
plt.xlabel('x-value', fontsize=16)
plt.ylabel('y-value', fontsize=16)
plt.grid()

plt.plot(data_x, yhat, c='y', lw=2, label='Regression Line') # c = color, lw = line width, label = name tag
 
plt.show()
````

#### B) Find the sum of squres of residuals about regrssion line

$$ S_r = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2  $$

````python
y = data_y
print(y)
print(yhat)

S_r = sum((y-yhat) ** 2)

print(f'S_r = {S_r}')
````

#### C) Find the standard error of estimate

$$ S_{y/x} = \sqrt{\dfrac{S_r}{n-2}} $$

````python
S_yx = np.sqrt(S_r/(n-2))

print(f'S_yx = {S_yx}')
````

#### D) Find sum of squares of residuals about the mean
$$ S_t = \sum_{i=1}^{n} (y_i - \bar{y})^2 $$

````python
print(y)
print(np.average(y))

ybar = np.average(y)
S_t = sum((y - ybar) ** 2)

print(f'S_t = {S_t}')
````

#### E) Find the coeﬀicient of determination
$$ r^2 = \dfrac{S_t - S_r}{S_t} $$

````python
r2 = (S_t - S_r)/S_t

print(f'r2 = {r2}')
````

#### F) Find the correlation coeﬀicient
$$ r = \sqrt{\dfrac{S_t - S_r}{S_t}}$$

````python
r = np.sqrt(r2)

print(f'r = {r}')
````

## Problem 02

The sales of a company (in million Baht) for each year are shown in the table below.

|  |  |  |  |  |  |
|---|------|------|------|-------|-------|
| $x$ (year) | 2016 | 2017 | 2018 | 2019 | 2020 |
| $y$ (sales) | 12 | 19 | 29 | 37  | 45  |

#### A) Find the least square regression line. 

````python
# find regression line
data_x = np.array([2016, 2017, 2018, 2019, 2020])
data_y = np.array([12, 19, 29, 37, 45])

n = data_x.shape[0]
ones = np.ones((n, 1))
X = np.concatenate((ones, data_x.reshape(n, 1)), axis = 1)
Y = data_y.reshape(n, 1)
B = np.linalg.inv(X.T @ X) @ X.T @ Y
print(f'yhat = {B[0]} + {B[1]} * x')

def model_p1(x):
    yhat = B[0] + B[1]*x
    return yhat

yhat = model_p1(data_x)

plt.scatter(data_x, data_y)
plt.grid()
plt.plot(data_x, yhat, c = 'y', lw = '2', label = 'Regression line')
plt.xlabel('Year', fontsize = 15)
plt.ylabel('Sales', fontsize = 15)
plt.show()
````

#### B) Find the coeﬀicient of determination

````python
# find r2 and r
y = data_y
S_r = sum((y - yhat) **2)
ybar = np.average(y)
S_t = sum((y - ybar) **2)
r2 = (S_t - S_r)/S_t
r = np.sqrt(r2)

print(f'r2 = {r2}')
print(f'r = {r}')
````

#### C) Use the least squares regression line as a model to estimate the sales of the company in 2021. 

````python
# predict sale of company in 2021
x = 2021
yhat = model_p1(x)
print(f'yhat = {yhat}')
````
